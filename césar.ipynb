# Cifrado César

```python
abecedario : str = 'abcdefghijklmnñopqrstuvwxyz'

def cifrar_letra(letra: str, desplazamiento: int) -> str:
    # Verificar si la letra es alfabética
    if letra.isalpha():
        # Determinar si la letra es mayúscula o minúscula
        es_mayuscula : bool = letra.isupper()
        # Convertir la letra a minúscula y encontrar su posición en el abecedario
        letra_minuscula : str = letra.lower()
        base: int = abecedario.index(letra_minuscula)
        nueva_posicion: int = (base + desplazamiento) % len(abecedario)
        # Obtener la letra cifrada y mantener si era mayúscula o minúscula
        letra_cifrada: str = abecedario[nueva_posicion]
        if es_mayuscula:
            return letra_cifrada.upper()
        else:
            return letra_cifrada
    else:
        return letra

def descifrar_letra(letra: str, desplazamiento: int) -> str:
    return cifrar_letra(letra, -desplazamiento)
```

```python
oracion = str(input("Ingrese una oración"))
desplazamiento = int(input("Ingrese un numero"))

for i in oracion:
  descifrar_letra(i, desplazamiento)


print(cifrar_letra(oracion, desplazamiento))
```
